```python
def get_public_holidays(year, country_code):
    url = f"https://date.nager.at/api/v3/PublicHolidays/{str(year)}/{country_code}"
    response = requests.get(url)
    if response.ok:
        return response.json()
    else:
        return f"Could not fetch holidays for {country_code} in {year}"

            

print(get_public_holidays(2020, "NO"))
print(get_public_holidays(2005, "NASFASD"))
```

```python
import datetime

def get_norwegian_holiday_dates(holiday_name, years_back):
    currentDate = datetime.date.today()
    year = currentDate.year

    holidays = []

    for i in range(years_back):
        public_holidays = get_public_holidays(year - i, "NO")
        for dict in public_holidays:
            if dict["localName"] == holiday_name:
                holidays.append(dict["date"])
            else:
                continue
    
    if holidays == []:
        return f"Could not find a match for {holiday_name}"
    else:
        return holidays
            
    

get_norwegian_holiday_dates("Langfredag",3)
```

```python
def get_fun_facts(list_of_dates):
    facts = {}
    for i in list_of_dates:
        date = i.split('-')
        response = requests.get(f"http://numbersapi.com/{date[1]}/{date[2]}/date")
        if response.ok:
            facts[i] = response.text
        else:
            return "Could not find facts for this date."
    return facts

get_fun_facts(get_norwegian_holiday_dates("Langfredag",3))
```
